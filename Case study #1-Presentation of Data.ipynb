```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1.Read the dataset to the Python environment

```python
cars_data=pd.read_csv('cars_data.csv')
cars_data.head()
```

```python
cars_data.dtypes
```

 Check for the nullvalues present in the dataset

3.Plot a Bargraph of male Vs female buyers participated in the sales

```python
plt.figure(figsize=(6,6))
sns.countplot(x='Buyer Gender',data=cars_data)
plt.grid()
plt.title('Gender wise participation in Sales',fontsize=15)
```

```python
Conclusion:
    1.From the Bargraph it is clear that participation of female is slightly higher than that of male in buying used cars.
```

4.Top 5 cars based on their Sales price
